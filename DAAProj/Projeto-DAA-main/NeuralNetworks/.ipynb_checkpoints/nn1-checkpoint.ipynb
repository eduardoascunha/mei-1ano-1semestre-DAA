```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
import os
import numpy as np
import random
```

```python
# tratamento da seed - pytorch
def set_seed(seed: int, deterministic_torch: bool = True):
    random.seed(seed) # Python
    np.random.seed(seed)  # Numpy, é o gerador utilizado pelo sklearn
    os.environ["PYTHONHASHSEED"] = str(seed)  # sistema operativo
    if torch is not None:
        torch.manual_seed(seed)
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)  #multi-GPU
        if deterministic_torch:
            torch.backends.cudnn.deterministic = True
            torch.backends.cudnn.benchmark = False

set_seed(25)
```

```python
class Model(nn.Module):
    def __init__(self, in_features=2161, h1=2000, h2=1500, h3=1000, h4=500, out_features=5):
        super().__init__()
        self.fc1 = nn.Linear(in_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.fc3 = nn.Linear(h2, h3)
        self.fc4 = nn.Linear(h3, h4)
        self.out = nn.Linear(h4, out_features)

    def foward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = self.out(x)

        return x
```

```python
torch.manual_seed(10)

model = Model()
```

```python
# Imports para manipulação e análise de dados
import pandas as pd
import numpy as np

# Imports para visualização
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
# Load dataset
control = pd.read_csv("../sbsppdaa24/train_radiomics_hipocamp.csv")

# Drop unique identifier columns
control.drop(columns=["Mask", "ID", "Image"], inplace=True)

# Drop non-numeric columns except for 'Transition'
columns_to_drop = [col for col in control.columns if control[col].dtype == 'object' and col != 'Transition']
control.drop(columns=columns_to_drop, inplace=True)

# Apply MinMax scaling to columns
from sklearn.preprocessing import MinMaxScaler
float_cols = control.select_dtypes(include=['float', 'int']).columns
scaler = MinMaxScaler()
control[float_cols] = scaler.fit_transform(control[float_cols])

# Aplicando Label Encoding para a variável 'Transition'
from sklearn.preprocessing import LabelEncoder
# Instanciando o LabelEncoder
encoder = LabelEncoder()

# Aplicando LabelEncoder diretamente na coluna 'Transition'
control['Transition'] = encoder.fit_transform(control['Transition'])

# Separando variáveis
X = control.drop(columns=['Transition'])
y = control['Transition']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2 ,random_state = 10)
```

```python
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)

y_train = torch.FloatTensor(y_train)
y_test = torch.FloatTensor(y_test)
```

```python
# Definir o dispositivo (CPU ou GPU) e mover o modelo e os dados para ele
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
X_train, X_test = X_train.to(device), X_test.to(device)
y_train, y_test = y_train.to(device), y_test.to(device)
```

```python
criterio = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
model.parameters
```

```python
print(X_train.shape)
```

```python
epochs = 110

losses = []

for i in range(epochs):
    y_pred = model.foward(X_train)

    #loss = criterio(y_pred, y_train) # predicted vs train
    loss = criterio(y_pred, y_train.long())
    
    #losses.append(loss.detach().numpy())
    losses.append(loss.detach().cpu().numpy())

    if i % 10 == 0:
        print(f"Epoch: {i} and loss: {loss}")

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plt.plot(range(epochs), np.array(losses))
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.show()
```

```python
criterion = nn.CrossEntropyLoss()

with torch.no_grad():
    y_eval = model.foward(X_test)
    loss = criterion(y_eval, y_test.long())
```

```python
loss # deve ser parecido com o ultimo loss do loop
```

```python
correct = 0
with torch.no_grad():
    for i, data in enumerate(X_test):
        y_val = model.foward(data)

        print(f"{i+1}.) {str(y_val)} \t\t {y_test[i]} \t\t {y_val.argmax().item()}")

        if y_val.argmax().item() == y_test[i]:
            correct += 1

print(f"corretas: {correct}")
```
