```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Load dataset
radi = pd.read_csv("sbsppdaa24/train_radiomics_hipocamp.csv")
radi.info()
```

```python
null_columns = radi.isnull().sum()
null_columns_with_values = null_columns[null_columns > 0]
null_columns_with_values
# Não há valores nulos
```

```python
duplicate_entries = radi.duplicated().sum()
duplicate_entries
# Não há entras duplicadas
```

```python
non_numeric_columns = radi.select_dtypes(exclude=[np.number]).columns.tolist()
len(non_numeric_columns),non_numeric_columns
# Não numércios
```

```python
constant_columns = [col for col in radi.columns if radi[col].nunique() == 1]
len(constant_columns)
# Colunas constantes
```

```python
# Display the count of each unique value in the 'Transition' column
transition_counts = radi["Transition"].value_counts()

# Plotting the bar graph
plt.figure(figsize=(8, 6))
transition_counts.plot(kind='bar', color='skyblue')
plt.title("Distribuição das entradas pelo target 'Transition'")
plt.xlabel("Transition")
plt.ylabel("Total")
plt.xticks(rotation=0)
plt.show()
```

```python
# Display summary statistics for each feature
summary_stats = radi.drop(columns=constant_columns,axis=1).describe().transpose()
summary_stats[['min', '25%', '50%', '75%', 'max']]
```

```python
# Identify constant columns (columns with only one unique value)
constant_columns = [col for col in radi.columns if radi[col].nunique() == 1]

# Identify non-numeric columns
non_numeric_columns = [col for col in radi.columns if radi[col].dtype == 'object' and col != 'Transition']

# Step 1: Perform One-Hot Encoding on the 'Transition' column
radi = pd.get_dummies(radi, columns=['Transition'], prefix='Transition')

# Step 2: Drop constant and non-numeric columns
filtered_radi = radi.drop(columns=constant_columns + non_numeric_columns)

# Step 3: Calculate correlation matrix
corr_matrix = filtered_radi.corr()

# Step 4: Identify top 10 features most correlated with each one-hot encoded target column
# List all one-hot encoded columns for 'Transition'
one_hot_columns = [col for col in filtered_radi.columns if col.startswith("Transition_")]

# Initialize a dictionary to store top correlated features for each encoded column
top_corr_features = {}

for col in one_hot_columns:
    # Get absolute correlations with the current one-hot encoded column
    target_corr = corr_matrix[col].drop(col).abs().sort_values(ascending=False)
    # Get top 10 most correlated features
    top_corr_features[col] = target_corr.head(10).index.tolist()

# Step 5: Create a filtered correlation matrix and plot heatmaps for each one-hot encoded target
for col in one_hot_columns:
    # Create filtered correlation matrix with top features and the target column
    top_features = top_corr_features[col]
    top_corr_matrix = corr_matrix.loc[top_features + [col], top_features + [col]]
    
    # Plot heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(top_corr_matrix, annot=True, cmap="coolwarm", cbar=True, square=True)
    plt.title(f"Heatmap of Top 10 Most Correlated Features with '{col}'")
    plt.show()
```

```python
import seaborn as sns

# Select only the first 50 features for box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=radi[summary_stats.index[:20]], orient="h", color='lightblue')
plt.title("Box Plot of First 50 Features")
plt.xlabel("Value")
plt.ylabel("Features")
plt.show()
```

```python
# Select only the first 50 features for the min and max range plot
plt.figure(figsize=(12, 8))
plt.plot(summary_stats['min'][:10], label='Min', marker='o', color='orange')
plt.plot(summary_stats['max'][:10], label='Max', marker='o', color='blue')
plt.fill_between(summary_stats.index[:10], summary_stats['min'][:10], summary_stats['max'][:10], color='lightgrey', alpha=0.5)
plt.title("Min and Max Values of First 50 Features")
plt.xlabel("Features")
plt.ylabel("Value")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
